## Analysis of the The Stanford Open Policing Project for Nashville, TN, spanning from March 2019 to December 2009.

Downlanded Nashville dataset. No renaming, then unzipped. 

Check https://github.com/stanford-policylab/opp/blob/master/data_readme.md for an detailed explanation of column names

```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns

from scipy import stats
```

```python
# the common datapath; unzipped name of the data file (not renamed)
data = "data/tn_nashville_2020_04_01.csv"

df = pd.read_csv(data)
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
sensitive_columns = ['location','subject_age', 'subject_race', 'subject_sex']
unique_identifiers = ['raw_row_number', 'officer_id_hash']

for column in unique_identifiers:
    if df[column].nunique() == len(df):
        print(f"Column '{column}' contains unique identifiers")

combined_data = df[sensitive_columns].drop_duplicates()
if len(combined_data) == len(df):
    print("Big yikes")
```
